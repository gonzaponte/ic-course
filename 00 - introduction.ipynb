# HDF5

All IC data files are stored in HDF5 format. These files work like a filesystem on their own. Groups behave like directories and nodes play the role of files. Each file has a tree of groups and nodes hanging from the root. To display an HDF5 file we will use the `ptdump` utility that comes with `pytables`.
- `ptdump    <filename>                  ` displays the file structure
- `ptdump -v <filename>                  ` displays the file structure and the node description
- `ptdump -d <filename>                  ` displays the file structure and dumps the data in each node
- `ptdump -options <filename>:/group     ` limits the output to the given group
- `ptdump -options <filename>:/group/node` limits the output to the given node

# IC files

All files processed by IC have some nodes that are always present:
- /Run/events: contains the event numbers and the timestamps of the events stored in this file
- /Run/runInfo: contains the run numbers associated to each event in the file (usually just one)
- /Run/eventMap: contains a mapping between the MC event number and the IC event number (more on this later)

At the same time, there are some special nodes. Some cities apply some filter to certain branches of the processing. The output of these filters is always stored to keep track of events. This is always written under the `/Filters` group as tables with two fields: the event number and a flag indicating whether the event passed. The name of the node is given by each filter in the city. For instance, in Sophronia we have two filters: `s12_selector` and `valid_hit`

# Executing IC

IC cities are run from the command line as `city <city_name> <config_file> [options]`. The options input can be used to override some of the parameters of the config file (see below).

## IC configuration

All IC cities have some common input arguments:
- `files_in`: input file or file list
- `file_out`: output file
- `run_number`: the run number
- `detector_db`: which detector configuration the data belongs to (new, next100, ...)
- `print_mod`: how often to print progress
- `compression`: compression library and compression level ("ZLIB4" is the most common one)
- `event_range:` range of events to process. Can be given in a number of ways:
  - `N`: process the first N events
  - `N, M`: skip N events, process the next M-N events
  - `N, last`: skip the first N events
  - `all`: process all events
  
In this course we will omit `files_in` and `file_out` from the config files and provide them from the command line (CLI)

## CLI overrides

Some of these can be overriden from the command line. For instance `city <city_name> -i <input_files> -o <output_file> -e <event_range>`:
- `-i <input_files>` will override the input files
- `-o <output_file>` will override the output files
- `-e <event_range>` will override the event range

## Run number

The run number in the config file implements a special behaviour in IC. The sensor information (position, status, calibration, etc.) can change over time and this is mapped through the run number. In order to process MC data assuming the sensor information from a certain run, we use the same number, but in negative, i.e.:
- `run_number =  N` means this is a file from the detector from run N
- `run_number = -N` means this is a file from MC processed as if it was from run N

The negative sign is also an indication that the MC information must be propagated from the input to the output file.

# Event number mapping

There is a subtle difference between what we call an event in nexus and in IC. In nexus, an event involves the entire physical process of a given phenomenon, regardless of how long it takes to complete. On the other hand, in IC, an event is limited to the window imposed by data acquisition. Therefore, a nexus event may be split into multiple IC events causing a mismatch between the event numbers. This mismatch is addressed by mapping IC event numbers to nexus event numbers. The exact difference is determined by the configuration of the cities `Buffy` or `Detsim`. The mapping is stored in the node `/Run/eventMap`:
